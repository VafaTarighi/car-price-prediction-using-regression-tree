```python
# constants

cities = [
    'tehran',
    'mashhad',
    'isfahan',
    'karaj',
    'shiraz',
    'tabriz',
    'qom',
    'ahvaz',
    'kermanshah',
    'urmia'
]
item = "car"
brand = "peugeot"
model = "206"
tipes = ['2', '5']
pages = ['0', '1', '2', '3', '4']

# search_api : "https://api.divar.ir/v8/web-search/{city}/{item}/{brand}/{model}/{type}"
```

```python

token_file = open(f'./{dataset_name}_car_tokens.txt', 'a')
token_list = []

for city in cities:
    for tipe in tipes:
        for page in pages:
            url = f"https://api.divar.ir/v8/web-search/{city}/{item}/{brand}/{model}/{tipe}?page={page}"
            resp = requests.get(url, headers={'Accept': 'application/json'})
            resp_json = resp.json()
            new_cars = resp_json['web_widgets']['post_list']
            new_tokens = list(map(lambda car : car['data']['action']['payload']['token'], new_cars))
            token_list.extend(new_tokens)
            for t in new_tokens:
                token_file.write(f'{t}\n')
            print(f"{city}:{tipe}type:{page}p : {len(new_tokens)} new car - total : {len(token_list)}")

token_file.write('\n')
token_file.close()
```

```python
collected = []
failed_tokens = []
```

```python
# pvm: persian_value_mapping
# features = city, brand, prod_year, color, usage, body_cond, engine_cond, front_chassis, back_chassis, tpi, gearbox, 'price'
token_list_size = len(token_list)
csv_file = open(f'./{dataset_name}_car_dataset.csv', 'a')
# write headers
csv_file.write('id,city,brand,prod_year,color,usage,body_cond,engine_cond,front_chassis,back_chassis,tpi,gearbox,price\n')


cars_dataset = []
i = 0
for token in token_list:
    if token in collected:
        i += 1
        print(f'skip {token} {i}/{token_list_size}')
        continue
    try:
        print(token, '')
        time.sleep(1.0)
        response = requests.get(f"https://api.divar.ir/v8/posts-v2/web/{token}", timeout=10)
        if response.status_code != 200:
            failed_tokens.append(token)
            i += 1
            print(f'timeout {token} {i}/{token_list_size}')
            continue

        car_json = response.json()


        car = {}
        if car_json.get('analytics') == None:
            print(car_json)
            if car_json['error'] == 'post not found':
                i += 1
                print(f'post not found - {token} {i}/{token_list_size}')
                continue
        else:
            car['city'] = car_json['analytics']['city']

        # find LIST_DATA section
        list_data = None 
        for sect in car_json['sections']:
            if sect['section_name'] == 'LIST_DATA':
                list_data = sect['widgets']
                break

        for widget in list_data:

            if widget['widget_type'] == 'GROUP_INFO_ROW':
                dats = widget['data']['items']
                for dat in dats:
                    if pfm[dat['title']] == 'usage':
                        car['usage'] = fa_num_to_eng(dat['value'])
                    elif pfm[dat['title']] == 'prod_year':
                        car['prod_year'] = fa_num_to_eng(dat['value'])
                    elif pfm[dat['title']] == 'color':
                        car['color'] = persian_color_map[dat['value']]

            elif widget['widget_type'] == 'UNEXPANDABLE_ROW':
                title, value = widget['data']['title'], widget['data']['value']
                if pfm[title] == 'ignore':
                    continue
                if pfm[title] in ['prod_year', 'usage', 'price']:
                    if pfm[title] == 'price':
                        if pvm.get(value) == 'agreement':
                            value = 'n/a'
                        elif pvm.get(value) == 'n/a':
                            value = 'n/a'
                        else:
                            value = fa_num_to_eng(value[:-6])
                    car[pfm[title]] = value
                elif pfm[title] == 'chassis_cond':
                    if pvm[value] == 'intact':
                        car['front_chassis'] = 'intact'
                        car['back_chassis'] = 'intact'
                    elif pvm[value] == 'both_impact':
                        car['front_chassis'] = 'impact'
                        car['back_chassis'] = 'impact'
                    elif pvm[value] == 'both_colored':
                        car['front_chassis'] = 'colored'
                        car['back_chassis'] = 'colored'
                else:
                    car[pfm[title]] = pvm[value]
        
        entry = f"{token},{car.get('city', 'n/a')},{car.get('brand', 'n/a')},{car.get('prod_year', 'n/a')},{car.get('color', 'n/a')},{car.get('usage', 'n/a')},{car.get('body_cond', 'n/a')},{car.get('engine_cond', 'n/a')},{car.get('front_chassis', 'n/a')},{car.get('back_chassis', 'n/a')},{car.get('tpi', 'n/a')},{car.get('gearbox', 'n/a')},{car.get('price', 'n/a')}"
        csv_file.write(f"{entry}\n")
        csv_file.flush()
        cars_dataset.append(car)
        i += 1
        print(f' - {entry}', '')
        print(f' - {i}/{len(token_list)}')
    except Exception as e:
        failed_tokens.append(token)
        print(e)

csv_file.close()
```
