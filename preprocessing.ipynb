```python
import pandas as pd
import numpy as np
```

```python
# load the csv file
df = pd.read_csv('./elim_car_dataset.csv')
df.drop(['id'], axis=1, inplace=True)
df
```

```python
df.info()
```

```python
def nan_stat():
    d_size = df.shape[0]
    for label in list(df):
        # count number of rows with missing values
        n_miss = df[label].isnull().sum()
        perc = n_miss / d_size * 100
        print('> %s, Missing: %d (%.1f%%)' % (label, n_miss, perc))

nan_stat()
```

```python
ctgf = ['city', 'brand', 'color', 'body_cond', 'engine_cond', 'front_chassis', 'back_chassis', 'gearbox']
ordf = ['prod_year', 'usage', 'tpi', 'price']

# Replace catagorical missing values with the most frequent values present in each column
df[ctgf] = df[ctgf].fillna(df[ctgf].mode().iloc[0])

# Replace ordinal missing values with the average values presented in each column
df[ordf] = df[ordf].fillna(df[ordf].median())
```

```python
# One Hot Encoding
ctgf = ['city', 'brand', 'color', 'body_cond', 'engine_cond', 'front_chassis', 'back_chassis', 'gearbox']
for catagory in ctgf:
    oh = pd.get_dummies(df[catagory], prefix=catagory)
    df = pd.concat([df, oh], axis=1)
    df.drop([catagory], axis=1, inplace=True)
```

```python
df.to_csv('elim_car_dataset_preprocessed.csv')
```
